# Data is Cleaned now, Lets Proceed for Visualization

# 1) Top 5 most Popular Artists

```python
top_5_popular_artists = df.groupby("artist").count().sort_values(by = "song_title" ,ascending = False)["song_title"][:5]
top_5_popular_artists
```

```python
# Visualization of the above Data

top_5_popular_artists.plot(kind='barh', figsize=(15,6),color="g")
plt.show()
```

# Among the Top 5 most Popular Artist "<u>Drake</u>" is in the first position which is very much liked by the subscribers of Spotify

# 2) Top 5 loudest tracks

```python
top_5_loudest_tracks = df[["loudness" , "song_title"]].sort_values(by = "loudness" , ascending = True)[:5]
top_5_loudest_tracks
```

```python
# Visualization of the above Data

plt.figure(figsize = (12,7))
sns.barplot(x = "loudness" , y = "song_title", data = top_5_loudest_tracks)
plt.title("Top 5 loudest tracks")
plt.show()
```

# "<u>Lyric Pieces, Book I Op. 12: I. Arietta</u>" has the loudest tracks i.e  <u>-33.097</u> 

# 3) Artist with the most danceability songs

```python
top_5_artists_danceability_song = df[["danceability" , "song_title" , "artist"]].sort_values(by = "danceability" , 
                                                                                             ascending = False)[:5]
top_5_artists_danceability_song
```

```python
# Visualization of the above Data

plt.figure(figsize = (12,7))
sns.barplot(x = "danceability" , y = "artist", data = top_5_artists_danceability_song)
plt.title("Top 5 Artists with Most Danceability Songs")
plt.show()
```

# "<u>Ben Remember</u>" is the first artist with the most danceability songs and then so on

# 4) Top 10 Instrumental tracks

```python
top_10_instrumental_songs = df[["instrumentalness" , "song_title" , "artist"]].sort_values(by = "instrumentalness" , 
                                                                                ascending = False)[:10]
top_10_instrumental_songs
```

```python
# Visualization of the above Data

plt.figure(figsize =(12,7))
plt.pie(x = "instrumentalness" , data = top_10_instrumental_songs , autopct = "%1.2f%%" ,
                                                                            labels = top_10_instrumental_songs.song_title)
plt.title("Top_10_Instrumental_Songs")
plt.show()
```

# Among the top 10 Instrumental Songs "<u>Senseless Order</u>" song_title is the highest i.e <u>10.17%</u>

# 5) Multiple Feature Plots
    * tempo
    * loudness
    * acousticness
    * danceability
    * duration_ms
    * energy
    * speechiness
    * valence

```python
Multiple_feature_plots = ['tempo','loudness','acousticness','danceability','duration_ms','energy','speechiness','valence']
```

```python
for feature_col in Multiple_feature_plots:
    x = df[df["target"] == 1][feature_col]
    y = df[df["target"] == 0][feature_col]
    
    plt.figure(figsize = (12,7))
    
    sns.distplot(x , bins = 30, label = "Positive" , color = "green")
    sns.distplot(y , bins = 30, label = "Negative" , color = "red")
    
    plt.legend(loc = "upper right")
    plt.title(f"Positive and Negative Histogram Plot for {feature_col}")
    plt.show()
```

# Positive and Negative Histogram Plot is plotted for different types of features present in the Dataset, based on whether the Songs are being liked by the Subscribers/People or not.

# Description of Multiple Feature Plots

* **<u>tempo</u>** = Tempo is the word we use to describe the speed at which we play a song.Between 130 - 140 ,the tempo is good and  liked by the listeners.

* **<u>loudness</u>** = The overall loudness of a track in decibels(dB). Loudness values are averaged across the entire track and arre useful for comparing relative loudness of tracks.Here around -7 ,People like the song and then it starts declining because it's going on extremely loud.

* **<u>acousticness</u>** = A song can be considered acoustic as long as an acoustic instrument, including voices, can be clearly heard. The value from 0 to 0.1 acoustic are mostly liked by the People and after that it is not liked by the People because it is not clearly audible.

* **<u>danceability</u>** = Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable. It describes how suitable a track is for dancing. Here we can see in the graph that a value of 0.7 is being displayed and it shows that the songs have the highest danceability.

* **<u>duration_ms</u>** = Duration of the song between 0.2 - 0.3 millliseconds is mostly liked and prefeered by the People.

* **<u>energ</u>**y = Song energy is a subtle quality that’s hard to define.Energy represents a perceptual measure of intensity and activity. Here energy between 0.6 - 0.9 is mostly liked by the People.

* **<u>speechiness</u>** = Speechiness detects the presence of spoken words in a track. Here the speechiness is between 0 - 0.1 which      most likely represent only music like tracks.

* **<u>valence</u>** = Valence describes the musical positiveness conveyed by a track. Tracks with high valence sound are more positive (e.g. happy, cheerful), while tracks with low valence sound are more negative (e.g. sad, depressed). Here the valence is between 0.2 to 0.8 which is shows that the People are loving positive songs.
